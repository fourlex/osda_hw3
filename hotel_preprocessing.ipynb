```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./data/hotel_booking.csv')
df['children'] = df['children'].fillna(0)
with pd.option_context('display.max_columns', None):
    display(df)
```

```python
categorical = [
    'deposit_type',
    'customer_type',
    'arrival_date_year',
    'market_segment',
    'arrival_date_month',
    'arrival_date_day_of_month',
    'meal',
    'hotel'
]

numeric = {
    'lead_time': [-0.01, 100, 200, 400, np.inf],
    'stays_in_week_nights': [-0.01, 3, np.inf],
    'total_of_special_requests': [-0.01, 1, np.inf],
    'days_in_waiting_list': [-0.01, 1, np.inf],
    'adults': [-0.01, 1, np.inf],
    'children': [-0.01, 0, np.inf],
    'babies': [-0.01, 0, np.inf],
    'booking_changes': [-0.01, 0, np.inf],
}
```

```python
df = df[categorical + list(numeric.keys()) + ['is_canceled']]

for col in categorical:
    print(col)
    binarized = pd.get_dummies(df[col], prefix=col)
    df = df.join(binarized).drop(columns=[col])

df = df.astype(float)
with pd.option_context('display.max_columns', None):
    display(df)
```

```python
hotel_X = df.drop(columns=['is_canceled'])
hotel_y = df['is_canceled']
hotel_X_train, hotel_X_test, hotel_y_train, hotel_y_test = train_test_split(
    hotel_X,
    hotel_y,
    train_size=700,
    test_size=300,
    random_state=0,
    stratify=hotel_y
)
hotel_X_train.to_csv('./data_preprocessed/hotel_X_train.csv', index=False)
hotel_X_test.to_csv('./data_preprocessed/hotel_X_test.csv', index=False)
hotel_y_train.to_csv('./data_preprocessed/hotel_y_train.csv', index=False)
hotel_y_test.to_csv('./data_preprocessed/hotel_y_test.csv', index=False)
```

```python
df_bin = df.copy()

for col, bins in numeric.items():
    print(col)
    binarized = pd.get_dummies(pd.cut(df[col], bins=bins), prefix=f'{col}')
    df_bin = df_bin.join(binarized).drop(columns=[col])

df_bin = df_bin.astype(int, errors='raise')

with pd.option_context('display.max_columns', None):
    display(df_bin)
```

```python
hotel_bin_X = df_bin.drop(columns=['is_canceled'])
hotel_bin_y = df_bin['is_canceled']
hotel_bin_X_train, hotel_bin_X_test, hotel_bin_y_train, hotel_bin_y_test = train_test_split(
    hotel_bin_X,
    hotel_bin_y,
    train_size=700,
    test_size=300,
    random_state=0,
    stratify=hotel_bin_y
)
hotel_bin_X_train.to_csv('./data_preprocessed/hotel_bin_X_train.csv', index=False)
hotel_bin_X_test.to_csv('./data_preprocessed/hotel_bin_X_test.csv', index=False)
hotel_bin_y_train.to_csv('./data_preprocessed/hotel_bin_y_train.csv', index=False)
hotel_bin_y_test.to_csv('./data_preprocessed/hotel_bin_y_test.csv', index=False)
```
