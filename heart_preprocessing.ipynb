```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./data/heart_disease_health_indicators_BRFSS2015.csv').dropna()
with pd.option_context('display.max_columns', None):
    display(df)
```

```python
categorical = [
    'Diabetes',
]

numeric = {
    'BMI': [-0.01, 20, 30, 40, np.inf],
    'PhysHlth': [-0.01, 0, 10, np.inf],
    'MentHlth': [-0.01, 0, 10, np.inf],
    'Age': [-0.01, 5, 10, np.inf],
    'Income': [-0.01, 3, 6, 7, np.inf],
    'Education': [-0.01, 3, 5, np.inf],
}

scores = [
    'GenHlth',
]
```

```python


df = df[categorical + list(numeric.keys()) + scores + ['HeartDiseaseorAttack']]

for col in categorical:
    print(col)
    binarized = pd.get_dummies(df[col], prefix=col)
    df = df.join(binarized).drop(columns=[col])

df = df.astype(float)
with pd.option_context('display.max_columns', None):
    display(df)
```

```python
heart_X = df.drop(columns=['HeartDiseaseorAttack'])
heart_y = df['HeartDiseaseorAttack']
heart_X_train, heart_X_test, heart_y_train, heart_y_test = train_test_split(
    heart_X,
    heart_y,
    train_size=700,
    test_size=300,
    random_state=0,
    stratify=heart_y
)
heart_X_train.to_csv('heart_X_train.csv', index=False)
heart_X_test.to_csv('heart_X_test.csv', index=False)
heart_y_train.to_csv('heart_y_train.csv', index=False)
heart_y_test.to_csv('heart_y_test.csv', index=False)
```

```python
df_bin = df.copy()

for col in scores:
    print(col)
    binarized = (df[[col]] >= 4).rename(columns={col: f'{col}_ge_4'})
    df_bin = df_bin.join(binarized).drop(columns=[col])

for col, bins in numeric.items():
    print(col)
    binarized = pd.get_dummies(pd.cut(df[col], bins=bins), prefix=f'{col}')
    df_bin = df_bin.join(binarized).drop(columns=[col])

df_bin = df_bin.astype(int, errors='raise')

with pd.option_context('display.max_columns', None):
    display(df_bin)
```

```python
heart_bin_X = df_bin.drop(columns=['HeartDiseaseorAttack'])
heart_bin_y = df_bin['HeartDiseaseorAttack']
heart_bin_X_train, heart_bin_X_test, heart_bin_y_train, heart_bin_y_test = train_test_split(
    heart_bin_X,
    heart_bin_y,
    train_size=700,
    test_size=300,
    random_state=0,
    stratify=heart_bin_y
)
heart_bin_X_train.to_csv('heart_bin_X_train.csv', index=False)
heart_bin_X_test.to_csv('heart_bin_X_test.csv', index=False)
heart_bin_y_train.to_csv('heart_bin_y_train.csv', index=False)
heart_bin_y_test.to_csv('heart_bin_y_test.csv', index=False)
```
